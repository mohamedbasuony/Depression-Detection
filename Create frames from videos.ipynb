```python
import cv2
import numpy as np
import glob
import os 
 
```

```python
path_of_the_directory= '/Users/mohamedbasuony/Desktop/Research/create videos'
x=0
y=[]
for foldername in os.listdir(path_of_the_directory):
    folder = os.path.join(path_of_the_directory, foldername)
    img_array = []
    print(folder)
    if folder == os.path.join(path_of_the_directory, ".DS_Store"):
        print("here?")
        os.remove(folder)
        continue
    for filename in os.listdir(folder):
        
        f= os.path.join(folder, filename)
        print(f)
        if f == os.path.join(folder, ".DS_Store"):
            print("here?")
            continue
        img = cv2.imread(f)
        print(img)
        height, width, layers = img.shape
        size = (width,height)
        img_array.append(img)
        x=x+1
    out = cv2.VideoWriter(f"{folder}.avi", 
                         cv2.VideoWriter_fourcc(*'MJPG'),
                         10, size)
    y.append(x)
    x=0
    for i in range(len(img_array)):
        out.write(img_array[i])
    out.release()
print(y)
```
