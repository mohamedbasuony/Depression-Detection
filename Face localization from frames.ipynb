```python
import cv2
import dlib
import numpy 
import os


PREDICTOR_PATH = "/Users/mohamedbasuony/Desktop/Research/shape_predictor_68_face_landmarks.dat"
predictor = dlib.shape_predictor(PREDICTOR_PATH)
detector = dlib.get_frontal_face_detector()


def TooManyFaces():
    return 1

def NoFaces():
    return 1

def get_landmarks(im):
    rects = detector(im, 1)

    if len(rects) > 1:
        return numpy.array(["no"])
        
    if len(rects) == 0:
        return numpy.array(["no"])

    return numpy.matrix([[p.x, p.y] for p in predictor(im, rects[0]).parts()])

def annotate_landmarks(im, landmarks):
    im = im.copy()
    for idx, point in enumerate(landmarks):
        pos = (point[0, 0], point[0, 1])
        cv2.putText(im, str(idx), pos,
                    fontFace=cv2.FONT_HERSHEY_SCRIPT_SIMPLEX,
                    fontScale=0.4,
                    
                    color=(0, 0, 255))
        cv2.circle(im, pos, 3, color=(0, 255, 255))
    return im

path_of_the_directory= '/Users/mohamedbasuony/Desktop/Research/save'
for foldername in os.listdir(path_of_the_directory):
    folder = os.path.join(path_of_the_directory, foldername)
    print(foldername)
    if foldername == ".DS_Store":
        continue
    for filename in os.listdir(folder):
        img= os.path.join(folder, filename)
        image = cv2.imread(img)
        if image is not None:
            print (img)
            landmarks = get_landmarks(image)
            if ("no" in landmarks):
                os.remove(img)
                continue
            image_with_landmarks = annotate_landmarks(image, landmarks)
            cv2.imwrite(img,image_with_landmarks)
            print ("Done")
        else:
            print ("failed")

```
