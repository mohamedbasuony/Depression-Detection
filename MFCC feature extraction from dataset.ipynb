```python
import os
import librosa
import pandas as pd
```

```python
# Set the directory containing the videos
video_dir = '/Users/mohamedbasuony/Desktop/Audio Features/Development/Freeform'

# Iterate over all the videos in the directory
for video_file in os.listdir(video_dir):
    # Check that the file is a video
    if video_file.endswith('.mp4'):
        # Extract the audio from the video
        y, sr = librosa.load(os.path.join(video_dir, video_file))

        # Split the audio into frames
        frames = librosa.util.frame(y, frame_length=1024, hop_length=512)

        # Extract MFCC features from each frame
        mfcc = librosa.feature.mfcc(y=y, sr=sr, n_mfcc=13)

        # Save the MFCC features to a CSV file
        df = pd.DataFrame(mfcc)
        df.to_csv(os.path.join(video_dir, video_file.replace('.mp4', '.csv')), index=False)
```
